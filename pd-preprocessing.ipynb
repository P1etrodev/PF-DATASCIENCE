```python

path ="archivos\metadata.json.gz"
import json
import gzip
def parse(path):
#Abrir gzip
    g = gzip.open(path, 'r')
    for l in g:
        yield eval(l)

#Convertir a  json estricto
def parse(path):
    g = gzip.open(path, 'r')
    for l in g:
        yield json.dumps(eval(l))
    f = open("output.strict", 'w')
    for l in parse("archivos\metadata.json.gz"):
        f.write(l + '\n')

#Pandas data frame
import pandas as pd
import gzip
def parse(path):
    g = gzip.open(path, 'rb')
    for l in g:
        yield eval(l)
def getDF(path):
    i = 0
    df = {}
    for d in parse(path):
        df[i] = d
        i += 1
    return pd.DataFrame.from_dict(df, orient='index')

df = getDF('archivos\metadata.json.gz')
```
